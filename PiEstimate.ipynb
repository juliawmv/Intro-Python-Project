```python
import random
import matplotlib.pyplot as plt
import numpy as np

# function taking in total number of points and outputing an estimate of pi
def pi(N):
    circle_sum = 0
    for i in range(N):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if x**2 + y**2 <= 1:
            circle_sum = circle_sum + 1
    pi_estimate = 4*(circle_sum/N)
    return pi_estimate

# setting the arrays for plotting
results = np.array(0)
Naxis = np.array(0)

# produces values for number of points vs pi estimate
for i in range(4):
    N = 10**i
    Naxis = np.append(Naxis,N)
    results = np.append(results,pi(N))

print(results)

plt.plot(Naxis,results)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# function taking in total number of points and outputing an estimate of pi
def pi(N):
    circle_sum = 0
    # generates points in the square and then goes through a statement to check if they are within the circle
    # if they are within the circle then it adds 1 to the points in the circle sum
    for i in range(N):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if x**2 + y**2 <= 1:
            circle_sum = circle_sum + 1
    # uses the amount of points in the circle to estimate pi
    pi_estimate = 4*(circle_sum/N)
    return pi_estimate

# setting the arrays for plotting
results = np.array(0)
Naxis = np.array(0)

# produces values for number of points vs pi estimate and adds them to arrays for plotting
for i in range(4):
    N = 10**i
    Naxis = np.append(Naxis,N)
    results = np.append(results,pi(N))

# test print result to make sure that the answers coming out of the above statement make sense
print(results)

# plot statement for number of points vs pi estimate result
plt.plot(Naxis,results)
plt.show()

# I think my function for estimating works pretty well. I didn't end up including higher total number of points
# because at 10000, the plot started to get hard to define. I can and have definitely produced a more accurate
# estimate of pi once it hit 100000 and 1000000, but I feel good with the number produced by N = 10000.
```
